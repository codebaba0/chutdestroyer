```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\max00\\Downloads\\ML programs\\study_data.csv")

# scatter plot of study hours vs exam scores
plt.figure(figsize= (14,7))
plt.subplot(1,2,1)
plt.scatter(data["Study Hours"],data["Exam Score"],color="blue",edgecolor="k",alpha=0.7)
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study hours")
plt.ylabel("Exam Scores")
plt.grid(True)

# bar chart of Average exam score by study hour range
bins = [0,2,4,6,8]
labels = ['0-2','2-4','4-6','6-8']
data['Study Hour Range']= pd.cut(data["Study Hours"], bins=bins,labels=labels,right=False)
grouped_data= data.groupby("Study Hour Range")["Exam Score"].mean()
plt.subplot(1,2,2)
grouped_data.plot(kind="bar",color="green")
plt.title("Average exam score by study hour range")
plt.xlabel("Study Hour Range")
plt.ylabel("Average Exam Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
