```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import export_text
import matplotlib.pyplot as plt

X=np.array([[150,0],[170,1],[200,0],[180,1],[220,1],[100,0]])
y=np.array(["Apple","Orange","Apple","Melon","Orange","Apple"])

clf=DecisionTreeClassifier(random_state=42)
clf.fit(X,y)

tree_rules= export_text(clf,feature_names=["Weight","Texture"])
print("Decision Tree Rules:\n",tree_rules)

plt.figure(figsize=[10,6])
plot_tree(clf,filled=True,feature_names=["Weight","Texture"], class_names=np.unique(y))
plt.show()
```
